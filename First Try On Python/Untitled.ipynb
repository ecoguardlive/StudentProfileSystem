```python

# ── 1. Collect student information ──────────
print("=" * 45)
print("       STUDENT PROFILE SYSTEM")
print("=" * 45)

name        = input("\nEnter your full name: ")
age         = int(input("Enter your age: "))
gpa         = float(input("Enter your GPA: "))
courses_raw = input("Enter courses enrolled (comma-separated): ")

# Convert comma-separated string to a list and strip whitespace
courses = [course.strip() for course in courses_raw.split(",")]

# ── 2. Store all data in a dictionary ───────
student_profile = {
    "name"       : name,
    "age"        : age,
    "gpa"        : gpa,
    "courses"    : courses,
    "num_courses": len(courses),
}

# ── 3. Display the formatted student profile ─
print("\n" + "=" * 45)
print("           STUDENT PROFILE")
print("=" * 45)
print(f"  Name        : {student_profile['name']}")
print(f"  Age         : {student_profile['age']} years old")
print(f"  GPA         : {student_profile['gpa']:.2f}")
print(f"  Courses     : {', '.join(student_profile['courses'])}")
print(f"  No. Courses : {student_profile['num_courses']}")
print("=" * 45)

# ── 4. Data type verification ────────────────
print("\n--- Data Types ---")
print(f"  name        → {type(student_profile['name'])}")
print(f"  age         → {type(student_profile['age'])}")
print(f"  gpa         → {type(student_profile['gpa'])}")
print(f"  courses     → {type(student_profile['courses'])}")
print(f"  num_courses → {type(student_profile['num_courses'])}")
```
